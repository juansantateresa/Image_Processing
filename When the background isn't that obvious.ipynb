```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage.filters import threshold_otsu
from skimage import data,color,io
```

```python
pape_image = io.imread('Image Processing with Python course exercise dataset/chapter 1/text_page_original.png')
io.imshow(pape_image)
```

```python
pape_image_gray = color.rgb2gray(pape_image)
```

```python
global_thres = threshold_otsu(pape_image_gray)
```

```python
binary_global = pape_image > global_thres
```

```python
io.imshow(binary_global)
```
