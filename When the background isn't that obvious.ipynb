```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage.filters import threshold_otsu
from skimage import data,color,io
from skimage.color import rgb2gray
```

```python
pape_image = plt.imread('Image Processing with Python course exercise dataset/chapter 1/text_page_original.png')
io.imshow(pape_image)
```

```python
pape_image_gray = rgb2gray(pape_image)
io.imshow(pape_image_gray)
```

```python
global_thres = threshold_otsu(pape_image_gray)
```

```python
binary_global = pape_image_gray > global_thres
```

```python
io.imshow(binary_global)
```

```python
from skimage.filters import threshold_local
```

```python
block_size = 35
```

```python
local_thres = threshold_local(pape_image_gray, block_size, offset = 10)
```

```python
pape_image
```

```python
binary_local = pape_image_gray > local_thres
```

```python
binary_local
```

```python
io.imshow(binary_local)
```
