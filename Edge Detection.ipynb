```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage import color, io, data
from skimage.color import rgb2gray
from skimage.filters import sobel
```

Let's extract the edge using sobel filter

```python
soaps_image = plt.imread('Image Processing with Python course exercise dataset/Chapter 2/soap_image.jpg')
```

```python
io.imshow(soaps_image)
```

```python
soaps_image_gray = rgb2gray(soaps_image)
```

```python
io.imshow(soaps_image_gray)
```

```python
edge_sobel = sobel(soaps_image_gray)
```

```python
io.imshow(edge_sobel)
```
