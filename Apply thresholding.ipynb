```python
import pandas as pd
import numpy as np
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
from skimage.filters import threshold_otsu
```

```python
from skimage import data, io
```

Let see which type of threshold should be applied

```python
tools_image = plt.imread('Image Processing with Python course exercise dataset/chapter 1/shapes52.jpg')
```

```python
io.imshow(tools_image)
```

```python
tools_image_gray = rgb2gray(tools_image)
```

```python
io.imshow(tools_image_gray)
```

```python
thresh = threshold_otsu(tools_image_gray)
```

```python
binary = tools_image_gray > thresh
```

```python
io.imshow(binary)
```

threshold_otsu is working fine
